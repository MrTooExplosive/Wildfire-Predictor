```python
def returnStationNumber(lat: float, long: float, county):
    long_list = []
    if lat != 0.0 and long != 0.0:
        max_lat_limit = lat
        min_lat_limit = trunc(max_lat_limit * 10) / 10
        test_df = station_list[station_list['Latitude'] <= max_lat_limit][station_list['Latitude'] >= min_lat_limit ]
        while test_df.empty and min_lat_limit > 0 and min_lat_limit < 1000:
#             print('Still in loop',min_lat_limit)
            min_lat_limit -= .1
            test_df = station_list[station_list['Latitude'] <= max_lat_limit][station_list['Latitude'] >= min_lat_limit ]
        if min_lat_limit < 0 or min_lat_limit > 1000:
            return 0.0
        for idx, row in test_df.iterrows():
            long_diff = row['Longitude'] - long
            long_list.append(long_diff)
    else:
        try:
            return station_list.loc[station_list['County'] == county]['Station Number'].iloc[0]
        except:
            return 0.0
    try:
        return test_df.loc[test_df['Longitude'] == max(long_list) + long, 'Station Number'].iloc[0]
    except:
        return 0.0
```
